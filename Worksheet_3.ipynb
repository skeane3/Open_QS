For $d=5$, numerically test your function: check the commutation relation $[\hat{a},\hat{a}^\dagger]$ and confirm that $\hat{a}|4\rangle$ yields the correct result, where $|4\rangle$ is a Fock state. 

```python
def fock_state(n, d):
    """
    Create the Fock state |n> in a d dimensional space

    :param int n: the number of the Fock state to be created
    :param int d: the dimension of the space in which the fock state will exist
    :return: ndarray representing the Fock state |n>
    """
    fock = np.zeros(d)
    fock[n] = 1
       
    return fock.reshape(-1, 1)

def annihilation(d):
    """
    Create the annihilation operator in d dimensions

    :param int d: the dimension of the space in which the fock state will exist
    
    :return: dxd ndarray representing the annihilation operator matrix
    """
    d = d
    a = np.zeros(d**2).reshape(d, d)
    for i in range(d):
        for j in range(d):
            if i-j == 1:
                a[j, i] = np.sqrt(i)
    
    return a

d = 5
a_5 = annihilation(d)
a_5_dagger = a_5.T

fock_4 = fock_state(4, d)
print(a_5@fock_4)

comm = a_5@a_5_dagger - a_5_dagger@a_5


print(comm)

```

Print its output for $\alpha = 0.5$ with $d=3$, $d=5$, and $d=10$. Comment on the difference between the results for different $d$. Is the truncated displacement operator unitary?

```python
def displacement(d, alpha):
    """
    Create a d dimenstional displacement operator based on the value aplha

    :param int d: the dimension of the space in which the fock state will exist
    :param complex alpha: The complex arguement of the Q function
    :return: dxd ndarray representing the displacement operator matrix
    """
    
    a = annihilation(d)
    a_dagger = a.T.conjugate()
    
    D = expm(alpha*a_dagger - alpha.conjugate()*a)
    
    return D

alpha = 0.5

d = 3
print(displacement(d, alpha))


d = 5
print(displacement(d, alpha))

d = 10
print(displacement(d, alpha))

# Is D unitary?

D = displacement(d, alpha)
D_dagger = D.T.conjugate()
print(np.round(D@D_dagger, 5))
```

```python
def pure_rho(phi):
    """
    Return the pure state density matrix representation of a state vector phi

    :param ndarray phi: a dx1 dimensional ndarray representing a state vector
    :return: ndarray representing the density matrix
    """
   
    # Normalization
    phi_bra = phi.T.conjugate()
    N = np.sqrt(phi_bra@phi)
    
    rho = (1/N)*(phi@phi_bra)
    return rho
    

def q_func(alpha, phi):
    
    """
    Return the value of the Q function for an arbitrary value of alpha and state rho
    This funciton uses a pure density matrix of the state rho

    :param complex alpha: The complex arguement of the Q function
    :param ndarray phi: a dx1 dimensional ndarray representing a state vector
    :return: complex value of the function
    """
        
    d = len(phi)
    # Setup the ground state
    ground_state = fock_state(0, d)
    
    # Setup the density matrix
    rho = pure_rho(phi)

    
    # Setup the displacement operator
    D = displacement(d, alpha)
    alpha_state = D@ground_state
    
    #Compute the Q-function
    Q = (1/np.pi)*(alpha_state.T.conjugate()@rho@alpha_state)
    
    return Q[0,0]

def plot_husimi(phi):
    
    """
    Return the value of the Q function for an arbitrary value of alpha and state rho
    This funciton uses a pure density matrix of the state rho

    :param complex alpha: The complex arguement of the Q function
    :param ndarray phi: a dx1 dimensional ndarray representing a state vector
    :return: complex value of the function
    """
    
    n = 100
    lim = 5
    
    xlist = np.linspace(-lim, lim, n)
    ylist = np.linspace(-lim, lim, n)
    X, Y = np.meshgrid(xlist, ylist)
    
    Z = np.zeros(len(xlist)*len(ylist)).reshape(len(ylist), len(xlist))
    for i in range(len(ylist)):
        for j in range(len(xlist)):
            Z[i, j] = np.real(q_func(xlist[j] + ylist[i]*1.0j, phi))

    fig = plt.figure(figsize=(10,8))
    ax = fig.add_axes([0.1,0.1,0.9,0.9])
    cp = ax.contourf(X, Y, Z)
    fig.colorbar(cp) # Add a colorbar to a plot
    ax.set_title('Filled Contours Plot')
    ax.set_xlabel('Re $alpha$')
    ax.set_ylabel('Im $alpha$')
    plt.show()
    
    

d = 30
fock_5 = fock_state(5, d)
plot_husimi(fock_5)
```

6. Let us now apply the truncated Hilbert-space representation to study the dynamics of the driven quantum harmonic oscillator, described by the Hamiltonian from question 2:
$$ \hat{H} = \hbar\omega_0\hat{a}^\dagger\hat{a} + \hbar F\cos(\omega_{\rm d} t)\left(\hat{a}+\hat{a}^\dagger\right).$$

(a) Write a function that generates this Hamiltonian (for any given $t$) in a truncated Hilbert space with $d=16$.

```python
def hamiltonian_b(t, w0, wd, F, d = 16, h_bar = 1):
    """
    Construct the above hamiltonian as a matrix for a given time t and in a given dimension

    :param float t: The time at which the Hamiltonin is evaluated
    :param d int: The dimensionality of the space on which the Hamiltonian is to act on
    :param w0 float: A constant associated with the Hamiltonioan
    :param wd float: A constant associated with the Hamiltonioan
    :param F float: A constant associated with the Hamiltonioan
    :return ndarray: A dxd ndarray representing the Hamiltoninan at time t
    """
    
    a = annihilation(d)
    
    return h_bar*w0*(a.T.conjugate()@a) + h_bar*F*np.cos(wd*t)*(a + a.T.conjugate())
    
```

(b) Numerically calculate the dynamics of the quantum state in the Schr√∂dinger picture starting from the vacuum state $|\psi(0)\rangle = |0\rangle$, and plot the mean value of the coordinate, $\langle \hat{q}(t)\rangle,$ as a function of time up to $\omega_0 t=100$.  Choose parameters $F=0.1\omega_0$ and $\omega_{\rm d} = 0.99\omega_0$.  (Recall that Worksheet 1 explored techniques to numerically calculate time evolution.)

```python
def evolve_b(phi_0, w0, wd, F, t_tot = 100, dt = 0.05):
    """
    Evolve an initial state phi_0 according to the above Hamiltonian

    :param ndarray: a 1xd vector representing the initial state of the system
    :param t_tot int: The total time over which the system is to be evolved (default = 100)
    :param dt float: The size of the time step increment (default = 0.05)
    :return ndarray: A ndarray containing the time value at each step in the evolution
    :return ndarray: A ndarray containing expectation value of the coordinate at each time step
    """
    
    # Setup steps and time array
    n_steps = int(t_tot/dt)  # Number of timesteps to take
    t_axis = np.zeros(n_steps+1) # Time axis (n_steps+1 includes t=0)  
    t = 0
    # The dimenstion of the system
    d = len(phi_0)
    
    # Setup the state phi and array to store the expectaiton values of the coordinate q
    phi = np.copy(phi_0)
    q_array = np.zeros(n_steps+1)
    
    # Initialise the coordinate operator q
    a = annihilation(d)
    a_dagger = a.T.conjugate()
    q = a + a_dagger
    
    for i in range(n_steps):
        
        t_axis[i] = t
        # Store the expectation value
        q_val = phi.T.conjugate()@q@phi
        q_array[i] = np.real(q_val[0, 0])
        # Calculate the evolution operator at time t
        U_dt = expm(-1j*dt*hamiltonian_b(t, w0, wd, F))
        # Update phi
        phi = U_dt@phi
        t += dt
    
    t_axis[-1] = t
    q_val = phi.T.conjugate()@q@phi
    q_array[-1] = np.real(q_val[0, 0])
    
        
    return t_axis, q_array
    
```

```python
# Set the associated constants as specified in the question
d = 16
w0 = 1
wd = 0.99*w0
F = 0.1*w0

# Initial ground state
phi_0 = fock_state(0, d)

t, q = evolve_b(phi_0, w0, wd, F)


fig = plt.figure(figsize=(10,8))
ax = fig.add_axes([0.1,0.1,0.9,0.9])
ax.plot(t, q)
ax.set_xlabel('$ w_0t$')
ax.set_ylabel('$< q >$')
ax.set_title('Expectation value of coordinate vs Time')
plt.show()
```

7. Another useful way to simulate *quadratic* Hamiltonians is to use the linearity of the Heisenberg equations. As in question 4, consider a two-mode system with annihilation operators $\hat{a}_1$ and $\hat{a}_2$. Convince yourself that the Hamiltonian can be written as a vector-matrix-vector product ([a quadratic form](https://en.wikipedia.org/wiki/Quadratic_form))
$$\hat{H}  = \mathbf{\hat{a}}^\dagger\cdot \mathbf{H} \cdot \mathbf{\hat{a}} = \sum_{j,k} H_{jk} \hat{a}_j^\dagger \hat{a}_k= 
\begin{pmatrix} \hat{a}_1^\dagger & \hat{a}_2^\dagger \end{pmatrix} \cdot \begin{pmatrix}
\hbar\omega_1  & \hbar g/2 \\ \hbar g/2 & \hbar\omega_2 \end{pmatrix}\cdot \begin{pmatrix} \hat{a}_1 \\ \hat{a}_2 \end{pmatrix},$$
where all three expressions above are just different ways of writing the same thing, with
$$ \mathbf{\hat{a}} = \begin{pmatrix} \hat{a}_1 \\ \hat{a}_2 \end{pmatrix},\qquad \mathbf{\hat{a}}^\dagger = \begin{pmatrix} \hat{a}_1^\dagger & \hat{a}_2^\dagger \end{pmatrix}, \qquad \mathbf{H} = \begin{pmatrix}
\hbar\omega_1  & \hbar g/2 \\ \hbar g/2 & \hbar\omega_2 \end{pmatrix},$$
while $H_{jk}$ are the elements of the $2\times 2$ Hermitian matrix $\mathbf{H}$.

(a) Consider the displacement vector $$\mathbf{d} = \begin{pmatrix}\langle \hat{a}_1 \rangle \\
\langle \hat{a}_2\rangle\end{pmatrix},$$
and the correlation matrix $$\mathbf{C} = \begin{pmatrix}\langle\!\langle \hat{a}_1^\dagger \hat{a}_1 \rangle\!\rangle & \langle\!\langle \hat{a}_2^\dagger \hat{a}_1 \rangle\!\rangle \\
\langle\!\langle \hat{a}_1^\dagger \hat{a}_2 \rangle\!\rangle &\langle\!\langle\hat{a}_2^\dagger \hat{a}_2 \rangle\!\rangle
\end{pmatrix},$$
or, equivalently, $d_j = \langle a_j\rangle$ and $C_{jk} = \langle\!\langle \hat{a}_k^\dagger \hat{a}_j\rangle\!\rangle = \langle \hat{a}_k^\dagger \hat{a}_j\rangle - \langle \hat{a}_k^\dagger\rangle\langle\hat{a}_j\rangle$ (note the order of indices $j,k$!). Use the Heisenberg equations to show that
$$ \frac{\rm d}{\rm dt}\mathbf{d} = \frac{1}{{\rm i}\hbar}\mathbf{H}\cdot \mathbf{d},\qquad  \frac{\rm d}{\rm dt} \mathbf{C} = \frac{1}{{\rm i}\hbar}[\mathbf{H},\mathbf{C}].$$ 

(b) Convince yourself that the solutions to the above equations are $$\mathbf{d}(t) = \mathbf{V}(t)\cdot\mathbf{d}(0),\qquad \mathbf{C}(t) = \mathbf{V}(t)\cdot\mathbf{C}(0)\cdot\mathbf{V}^\dagger(t),$$ where $\mathbf{V}(t) = {\rm e}^{-{\rm i}\mathbf{H} t/\hbar}$ is a unitary matrix. Therefore, the dynamics can be simulated using the same iterative method as we have previously used for qubits in Worksheet 1, since $$\mathbf{d}(t +\Delta t) = \mathbf{V}(\Delta t) \cdot \mathbf{d}(t), \qquad \mathbf{C}(t+\Delta t) = \mathbf{V}(\Delta t)\cdot\mathbf{C}(t)\cdot\mathbf{V}^\dagger(\Delta t).$$ Use this method to calculate the dynamics of the system numerically, starting from the product state 
$$|\psi(0)\rangle = |4\rangle \otimes |0\rangle,$$
corresponding to a state with 4 energy quanta in mode 1 and 0 quanta in mode 2. Generate plots of the mean coordinate $\langle \hat{q}_j(t)\rangle$ and the mean occupation number $\langle \hat{n}_j(t)\rangle$ for both modes, $j=1,2$. Choose parameters $\omega_2=0.9\omega_1$, $g=0.3\omega_1$, and evolve up to a time $\omega_1t = 50$.

```python
def exp_val(operator, phi):
    """
    Calculate the expectation of given operator with respect to some state phi

    :param operator ndarray: nxn ndarray representing the matrix form of an operator
    :param phi ndarray: A 1xn ndarray representing the state of the system
    :return complex: A complex number representing the expectation value of the operator
    """
    return ((phi.T.conjugate())@(operator@phi))[0][0]

# Define a function for iteratively carrying out unitary time evolution, with U_dt the evolution operator over
# one timestep, initial state psi_0, total evolution time t_tot, timestep dt, and obs a list of observables
def time_evo(phi_0, w1, w2, g, dim, t_tot, dt, q_0 = 1):
    """
    Evolve an initial state phi_0 in time according to the Hamiltonian in question 7

    :param phi_0 ndarray: A 1xn ndarray representing the initial state of the system
    :param w1 float: A constant associated with the Hamiltonioan
    :param w2 float: A constant associated with the Hamiltonioan
    :param g float: A constant associated with the Hamiltonioan
    :param dim int: The dimensionality of the space on which the Hamiltonian is to act on
    :param t_tot float: The total time over which the system is evolved
    :param dt float: The time step increment that the system will be evolved by
    :param q0 float: A constant associated with the the position operator (default = 1)
    
    :return ndarray t_axis: containing the time value at each increment in the evolution
    :return ndarray q: containing the position expectation value of each mode at each time increment t
    :return ndarray n: containing the number expectation value of each mode at each time increment t
    """
    # Initialise our H matrix and evolution operator V
    H = np.array([[w1, g/2], [g/2, w2]])
    V_dt = expm(-1j*H*dt)
    V_dt_dagger = V_dt.T.conjugate()
    
    # Initialise the anhiallation operator
    a = annihilation(dim)
    
    # Initialise the ladder operators to act on each mode
    a = annihilation(dim)
    a1 = np.kron(a.copy(), np.eye(dim))
    a2 = np.kron(np.eye(dim), a.copy())
        
    a1_dagger = a1.T.conjugate()
    a2_dagger = a2.T.conjugate()

        
    # Initialise our d and C matrices
    d = np.array([[exp_val(a1, phi_0)], [exp_val(a2, phi_0)]])
      


    C = np.array([[exp_val(a1_dagger@a1, phi_0), exp_val(a2_dagger@a1, phi_0)],
                  [exp_val(a1_dagger@a2, phi_0), exp_val(a2_dagger@a2, phi_0)]])

    
    # Initialise variables associeated with the evolution
    n_steps = int(t_tot/dt)  # Number of timesteps to take
    t_axis = np.zeros(n_steps+1) # Time axis (n_steps+1 includes t=0)  
    # Time value at any given moment in the evolution
    t = 0
    # Initialise arrays to store the expectation values of each mode of 
    # q and n at each time step
    q = np.zeros(2*(n_steps+1)).reshape(2, n_steps+1)
    n = np.zeros(2*(n_steps+1)).reshape(2, n_steps+1)
    # Loop over timesteps
    
    for i in range(n_steps):
        
        # Compute and store expectation values
        t_axis[i] = t
        for j in range(2):
            q[j, i] = np.real(q_0*(d[j, 0] + d[j, 0].conjugate()))
            n[j, i] = np.real(C[j, j] + (d[j, 0].conjugate())*d[j, 0])
        
        # Take a step forward in time
        d = V_dt@d
        C = V_dt@C@V_dt_dagger
        t += dt
    
    # Compute and store final expectation values
    t_axis[-1] = t
    for j in range(2):
        q[j, -1] = np.real(q_0*(d[j, 0] + d[j, 0].conjugate()))
        n[j, -1] = np.real(C[j, j] + (d[j, 0].conjugate())*d[j, 0])
    
    return t_axis, q, n
    
```

(c) Let us now apply these results to simulating the dynamics of a single oscillator system in contact with a large many-mode bath. Consider the Hamiltonian $\hat{H} = \hat{H}_S + \hat{H}_B + \hat{H}_{SB}$, where
\begin{align}
&\hat{H}_S = \hbar\nu_0 \hat{a}_0^\dagger \hat{a}_0 \\,\qquad &\hat{H}_B = \sum_{j=1}^{M-1} \hbar\nu_j \hat{a}_j^\dagger \hat{a}_j + \sum_{j=1}^{M-2} \frac{\hbar W}{2} \left( \hat{a}_j^\dagger\hat{a}_{j+1} + \hat{a}_j\hat{a}_{j+1}^\dagger \right),\\
& \hat{H}_{SB} = \frac{\sqrt{\Gamma W}}{2}\left(\hat{a}_0^\dagger \hat{a}_1 + \hat{a}_0\hat{a}_1^\dagger\right).
\end{align}
This models a bath comprising a one-dimensional chain of oscillators coupled to the system at one of its ends ($j=1$). Write a function that constructs the matrix $\mathbf{H}$ for the *full* system-bath Hamiltonian $\hat{H} = \sum_{j,k}H_{jk}\hat{a}_j^\dagger\hat{a}_k$, taking the parameters $\{\nu_j\}$, $W$, $\Gamma$, and $M$ as input. Write another function that constructs the corresponding thermal correlation matrix $\mathbf{C}$. Test your function by printing the thermal correlation matrix for the parameters $M=4$, $\nu_j = \nu_0 = 2$, $W=0.5$, $\Gamma=0.05$ and $\beta=0.5$.

```python
def full_H(v, W, G, M):
    
    """
    Construct the matrix form of the above H matrix


    :param v ndarray: A 1xM array of mode energies associated with the Hamiltonioan
    :param W float: A constant associated with the Hamiltonioan, which couples adjacent modes
    :param G float: A constant associated with the Hamiltonioan, which couples the system to the bath
    :param beta float: A constant associated with the Hamiltonioan, inversley proportional to temperature
    :return ndarray: A dxd ndarray representing the Hamiltoninan at time t
    """
    
    # Initialise an empty ndarray for H
    H = np.zeros(M**2).reshape(M, M)
    
    # First row and first column entries
    H[0, 0] = v[0]
    H[0, 1] = G*W/2
    H[1, 0] = G*W/2
    
    # Middle entries
    for i in range(1, M-1):
        H[i, i] = v[i]
        H[i, i+1] = W/2
        H[i+1, i] = W/2
        
    # Last entry
    H[-1, -1] = v[-1]
    
    return H

def correlation_matrix(v, W, G, M, beta, H = None):
    """
    Construct the correlation matrix assuming a thermal state


    :param v ndarray: A 1xM array of mode energies associated with the Hamiltonioan
    :param W float: A constant associated with the Hamiltonioan, which couples adjacent modes
    :param G float: A constant associated with the Hamiltonioan, which couples the system to the bath
    :param beta float: A constant associated with the Hamiltonioan, inversley proportional to temperature
    :param H ndarray: A MxM array representing the Hamiltonian of the system
    :return ndarray: A MxM ndarray representing correlation matrix
    """
    # If no Hamiltonian is provided, then use H_full as default
    if not isinstance(H, np.ndarray):
        H = full_H(v, W, G, M)
    C = 1/(expm(beta*H) - 1)
    
    return C

# Initialise parameters
M = 4
W = 0.5
G = 0.05
beta = 0.5
v_0 = 2
v = np.array([v_0 for i in range(M)])

C = correlation_matrix(v, W, G, M, beta)
print(C)
```

(d) Now let us perform a dynamical simulation of a thermalisation process. Start from the product state $$\hat{\rho}(0) = |0\rangle\langle 0|\otimes \frac{{\rm e}^{-\beta\hat{H}_B}}{Z_B},$$

Construct the correlation matrix for this state and then evolve it as a function of time using the method of question 7, with parameters $\nu_j =\nu_0 = 2$, $W=0.5$, $\Gamma=0.05$, $\beta=0.5$, $M=51$, and up to a time $\Gamma t = 10$.  Specifically, plot the occupation of the system oscillator, $\langle \hat{a}_0^\dagger \hat{a}_0\rangle$, as a function of time. Explore how this open-system evolution changes as the number of oscillators, $M$, and the system-reservoir coupling, $\Gamma$, is increased or decreased. Comment briefly on your results.   


```python
def exp_val(operator, rho):
    """
    Calculate the expectation of given operator with respect to some density matrix rho

    :param operator ndarray: nxn ndarray representing the matrix form of an operator
    :param rho ndarray: A nxn ndarray representing the density matrix of the system
    :return complex: A complex number representing the expectation value of the operator
    """
        
    return np.trace(rho@operator)

# Define a function for iteratively carrying out unitary time evolution, with U_dt the evolution operator over
# one timestep, initial state psi_0, total evolution time t_tot, timestep dt, and obs a list of observables
def time_evo_8c(v, W, G, M, beta, t_tot = 10, dt = 0.05):
    """
    Evolve an initial state phi_0 in time according to the Hamiltonian in question 7

    :param v ndarray: A 1xM array of mode energies associated with the Hamiltonioan
    :param W float: A constant associated with the Hamiltonioan, which couples adjacent modes
    :param G float: A constant associated with the Hamiltonioan, which couples the system to the bath
    :param beta float: A constant associated with the Hamiltonioan, inversley proportional to temperature
    :param t_tot float: The total time over which the system is evolved
    :param dt float: The time step increment that the system will be evolved by
    :param q0 float: A constant associated with the the position operator (default = 1)
    
    :return ndarray t_axis: containing the time value at each increment in the evolution
    :return ndarray q: containing the position expectation value of each mode at each time increment t
    :return ndarray n: containing the number expectation value of each mode at each time increment t
    """
    
    # Initialise the Hamiltonian of the system
    H = full_H(v, W, G, M)
    # Set the evolution operator to evolve the correlation matrix
    V_dt = expm(-1j*H*dt)
    V_dt_dagger = V_dt.T.conjugate()
    
    # Since the first row and column are associated with the system
    # Get the H matrix associated with the bath
    H_B = H[1:, 1:]
    
    # Initialise an array to store the correlation matrix
    C = np.zeros(M**2).reshape(M, M)
    
    # The first row and coulumn associated with the system will be zero as it is 
    # initailly in the vacuum state. Set the rest of the entries to the correlation 
    # matrix of the thermal state
    C[1:, 1:] = correlation_matrix(v, W, G, M-1, beta, H_B)
    
    # Initialise variables
    n_steps = int(t_tot/dt)  # Number of timesteps to take
    t_axis = np.zeros(n_steps+1) # Time axis (n_steps+1 includes t=0)  
    t = 0
    # Initialise arrays to store the expectation values of 
    # n at each time step
    n = np.zeros((n_steps+1))
    # Loop over timesteps
    
    for i in range(n_steps):
        
        # Compute and store expectation values
        t_axis[i] = t
        
        n[i] = np.real(C[0, 0])
        
        # Take a step forward in time

        C = V_dt@C@V_dt_dagger
        t += dt
    
    # Compute and store final expectation values
    t_axis[i+1] = t
  
    n[-1] = np.real(C[0, 0])
    
    return t_axis, n
```

Notice that $\langle \hat{a}_0^\dagger \hat{a}_0\rangle$ is the the expectation value of the number operator of the system. This means that it is essentially countinf the number of number of particles preent in our system. Since our system starts from the ground state we see that t at time $t = 0$, the expectation value is 0. As time progresses this expectation value slowly increases as energy is absorbed form the bath which creates particles within our system.  
We can see that increasing the number of modes in the bath does not change the expectation value of the number operator within the system. Increasing the number of modes means that there are more modes to transfer energy to the sytsem but it also means that the bath energy is further distributed amongst the nodes. Since the bath is in thermal equilibrium, increased transfer potential due the extra nodes is counteracted by the wider distribution resultng in no change in the expectation value.  
Increasing $\Gamma$ on the other hand has the effect of increasing the coupling between the system and the bath. This means the the system and the bath interact more strongly and thus energy can be transfered much faster between the two. The result is that we see the average number of particles in the system increase and a much faster rate as the coupling constant increases.
